```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df1=pd.read_csv('dataset1.csv')
df2=pd.read_csv('dataset2.csv')
df3=pd.read_csv('dataset3.csv')
```

```python
def preprocess(df):
  col_to_remove = []
  for col in df.columns:
    if col.startswith('Unnamed:') or col=='index': 
      col_to_remove.append(col)
  if 'STDs:AIDS' in list(df.columns): 
    df.drop(['STDs:AIDS','STDs:cervical condylomatosis'],axis=1, inplace=True)

  df.drop(col_to_remove,axis=1, inplace=True)
  return df
```

```python
df1=preprocess(df1)
df2=preprocess(df2)
df3=preprocess(df3)
```

```python
def plotNone(df):
  x=[]
  y=[]
  for col in df.columns:
    column=df[col].tolist()
    cnt=max(column.count('?'),column.count('unknown'))
    if cnt>0:
      x.append(col)
      y.append(100*cnt/df.shape[0])
  plt.plot(x, y,marker='.')
  plt.xlabel('Column')
  plt.xticks(rotation=90)
  plt.ylabel('% of Null Values')
  plt.show()  
```

```python
plotNone(df1)
```

```python
plotNone(df3)
```

```python
def outlierCheck(df):
  columns=list(df.columns)
  target=columns[-1]
  columns=columns[:len(columns)-1]
  print(target)
  for col in df.columns:
    _, vars = plt.subplots()
    vars.scatter(df[col], df[target])
    vars.set_xlabel(col)
    vars.set_ylabel(target)
    plt.xticks(rotation=90)
    plt.show()
  print('<-------------------End of Dataset------------------->')
```

```python
outlierCheck(df1)
outlierCheck(df2)
outlierCheck(df3)
```

```python
def heatMap(df):
  col_to_remove = []
  for col in df.columns:
    if col.startswith('Unnamed:') or col=='index': 
      col_to_remove.append(col)
  df.drop(col_to_remove,axis=1, inplace=True)
  df=df.replace('?',np.nan)
  df=df.replace('unknown',np.nan)
  print(df.dtypes)
  df.fillna(df.median(), inplace=True)
  for col in df.columns:
    if df[col].dtype==object and 'Smokes' in list(df.columns):
      df[col]=df[col].astype(float)
    elif df[col].dtype==object and 'job' in list(df.columns):
      df[col]=df[col].astype(str)
  if 'job' in list(df.columns):
    df['job'].fillna(df['job'].mode()[0],inplace =True)
    df['marital'].fillna(df['marital'].mode()[0],inplace =True)
    df['education'].fillna(df['education'].mode()[0],inplace =True)
    df['housing'].fillna(df['housing'].mode()[0],inplace =True)
    df['loan'].fillna(df['loan'].mode()[0],inplace =True)
    df['default'].fillna(df['default'].mode()[0],inplace =True)
    df['job']=LabelEncoder().fit_transform(df['job'].astype('str'))
    df['marital']=LabelEncoder().fit_transform(df['marital'].astype('str'))
    df['education']=LabelEncoder().fit_transform(df['education'].astype('str'))
    df['housing']=LabelEncoder().fit_transform(df['housing'].astype('str'))
    df['loan']=LabelEncoder().fit_transform(df['loan'].astype('str'))
  # df = df.dropna()
  # print(df.shape)
  # print(df.corr())
  sns.heatmap(df.corr(),annot=True, xticklabels=True, yticklabels=True)
  plt.gcf().set_size_inches(25, 10)
  plt.show()
```

```python
heatMap(df1)
heatMap(df2)
heatMap(df3)
```
