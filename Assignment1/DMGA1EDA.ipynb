```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('dataset1.csv')
df2=pd.read_csv('dataset2.csv')
df3=pd.read_csv('dataset3.csv')
```

```python
def preprocess(df):
  col_to_remove = []
  for col in df.columns:
    if col.startswith('Unnamed:'): 
      col_to_remove.append(col)
  df.drop(col_to_remove,axis=1, inplace=True)
  return df
```

```python
df1=preprocess(df1)
df2=preprocess(df2)
df3=preprocess(df3)
```

```python
def findVals(df):
  y=[]
  for col in df.columns:
    print(col,'---->',df[col].unique())
    print()
```

```python
findVals(df2)
```

```python
findVals(df3)
```

```python
def plotNone(df):
  x=[]
  y=[]
  for col in df.columns:
    column=df[col].tolist()
    cnt=max(column.count('?'),column.count('unknown'))
    if cnt>0:
      x.append(col)
      y.append(100*cnt/df.shape[0])
  plt.plot(x, y,marker='.')
  plt.xlabel('Column')
  plt.xticks(rotation=90)
  plt.ylabel('% of Null Values')
  plt.show()  
```

```python
plotNone(df1)
```

```python
plotNone(df3)
```

```python
def outlierCheck(df):
  columns=list(df.columns)
  target=columns[-1]
  columns=columns[:len(columns)-1]
  print(target)
  for col in df.columns:
    _, vars = plt.subplots()
    vars.scatter(df[col], df[target])
    vars.set_xlabel(col)
    vars.set_ylabel(target)
    plt.xticks(rotation=90)
    plt.show()
  print('<-------------------End of Dataset------------------->')
```

```python
outlierCheck(df1)
outlierCheck(df2)
outlierCheck(df3)
```

```python
def heatMap(df):
  col_to_remove = []
  for col in df.columns:
    if col.startswith('Unnamed:') or col=='index': 
      col_to_remove.append(col)
  df.drop(col_to_remove,axis=1, inplace=True)
  df=df.replace('?',np.nan)
  df=df.replace('unknown',np.nan)
  # print(df.columns)
  # df = df.dropna()
  print(df.shape)
  sns.heatmap(df.corr(),annot=True)
  plt.gcf().set_size_inches(15, 10)
  plt.show()
```

```python
heatMap(df1)
heatMap(df2)
heatMap(df3)
```
